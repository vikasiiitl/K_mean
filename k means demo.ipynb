```python
from sklearn.cluster import KMeans
import pandas as pd
```

```python
from sklearn import datasets
iris=datasets.load_iris()
X=iris.data[:,:2]
y=iris.target
print(X[0:5])
print(y[0:5])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(X_train)
print("No of training sample:",len(X_train))
predict_train=kmeans.predict(X_train)
print(kmeans.predict(X_train))
```

```python
y_kmeans=kmeans.predict(X_test)
print("Predicted table:",y_kmeans)
print("No of testing data:",len(X_test))
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(X_train)
print("No of training sample:",len(X_train))
print(kmeans.predict(X_train))
```

```python
y_kmeans=kmeans.predict(X_test)
print("Predicted table:",y_kmeans)
print("No of testing data:",len(X_test))
```

```python
from sklearn.metrics import silhouette_score
print("K Means Silhouette Score",(silhouette_score(X_train,predict_train)))
```

```python
from sklearn.metrics.cluster import adjusted_rand_score
print("ARI of testing data",adjusted_rand_score(y_test,y_kmeans))
```
