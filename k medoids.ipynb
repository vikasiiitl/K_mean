```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets   #some datasets are already avaiable in sklearn library
iris = datasets.load_iris()
#print(iris)
iris_df=pd.DataFrame({'x': iris.data[:,0], 'y': iris.data[:,1], 'z': iris.data[:,2], 'm': iris.data[:,3]})   # we only take the first two features.
print(iris_df.head()) #printing top 5 rows of dataframe
#print(iris_df)
```

```python
kmedoids = KMedoids(n_clusters=3)
kmedoids.fit(iris_df)
y_kmediods = kmedoids.predict(iris_df)
print(y_kmediods)
```

```python
from sklearn.metrics import silhouette_score
print("Silhouette Score(n=3):", silhouette_score(iris_df, y_kmediods))
```

```python
centers=kmedoids.cluster_centers_
print(centers)
```

```python
plt.scatter(iris_df['x'],iris_df['y'],c=y_kmediods,s=50,cmap='viridis')
centers=kmedoids.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black',s=200);
plt.xlabel('Feature-1')
plt.ylabel('Feature-2')

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets   #some datasets are already avaiable in sklearn library
iris = datasets.load_iris()
#print(iris)
iris_df=pd.DataFrame({'x': iris.data[:,0], 'y': iris.data[:,1], 'z': iris.data[:,2], 'm': iris.data[:,3]})   # we only take the first two features.
print(iris_df.head()) #printing top 5 rows of dataframe
#print(iris_df)
```

```python
from sklearn.cluster import KMeans  #Importing K-means algorithm
kmeans = KMeans(n_clusters=3)
kmeans.fit(iris_df)
y_kmeans = kmeans.predict(iris_df)  #predict the labels of samples
print(y_kmeans)
```

```python
print("Silhouette Score(n=3):", silhouette_score(iris_df, y_kmeans))
```
